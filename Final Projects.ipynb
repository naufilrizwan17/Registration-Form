```python
import tkinter as tk
from tkinter import*
reg = tk.Tk()
reg.title("Registration Form")
reg.geometry('500x300')
reg.configure(bg = "white")

canvas_1 = tk.Canvas(reg,width = 800,height = 500)
canvas_1.pack()

label_1 = tk.Label(reg,text = "Registration Form")
label_1.configure(bg = "white",fg = "black",font = ("arial", 25,'bold'))
label_1.place(x = 15,y = 10)

label_2 = tk.Label(reg,text = "Full Name")
label_2.configure(bg = "white",fg = "black",font = ("arial", 10))
label_2.place(x = 15,y = 60)
 
entry_1 = Entry(reg)
entry_1.place(x = 100,y = 60)

label_3 = tk.Label(reg,text = "Email")
label_3.configure(bg = "white",fg = "black",font = ("arial", 10))
label_3.place(x = 15,y = 110)    

entry_2 = Entry(reg)
entry_2.place(x = 100,y = 110)

label_4 = tk.Label(reg,text = "Gender")
label_4.configure(bg = "white",fg = "black",font = ("arial", 10))
label_4.place(x = 15,y = 160)


b = IntVar()

Radiobutton(reg,text = "Male",padx = 5,variable = b,value = 1).place(x = 100,y = 160)
Radiobutton(reg,text = "Female",padx = 5,variable = b,value = 2).place(x = 150,y = 160)
                  

label_5 = tk.Label(reg,text = "Country")
label_5.configure(bg = "white",fg = "black",font = ("arial", 10))
label_5.place(x = 15,y = 210)
                  
list_of_country = ["Australia","Afghanistan","Belgium","Bangladesh","Brazil","England","Germany","Iraq","Iran","India","Pakistan","Portugal","Russia","Spain"]        

                  
c = StringVar()
                  
                  
droplist = OptionMenu(reg, c, *list_of_country)
droplist.configure(width = 15)
c.set('select your country')
droplist.place(x= 100,y= 210)
                  
                  

label_6 = tk.Label(reg,text = "Programming")
label_6.configure(bg = "white",fg = "black",font = ("arial", 10))
label_6.place(x = 15,y = 260)
                  
d = IntVar()
                  
Checkbutton(reg,text = "Java",variable = d).place(x = 100,y = 260)

e = IntVar()
                  
Checkbutton(reg,text = "Python",variable = e).place(x = 150,y = 260)
                  
                  
Button(reg,text = "Submit",bg = "red",width = 15).place(x = 15,y = 310)
                  
reg.mainloop()                  
```
